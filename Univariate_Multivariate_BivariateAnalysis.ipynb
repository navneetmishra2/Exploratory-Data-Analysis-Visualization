```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=1000, n_features=6, n_informative=4, n_redundant=2, 
                           n_classes=3, random_state=42)

# Convert to DataFrame
columns = ["Age", "Salary", "Experience", "WorkingHours", "Projects", "PerformanceScore"]
df = pd.DataFrame(X, columns=columns)

# Add categorical columns
df["Department"] = np.random.choice(["HR", "IT", "Finance", "Marketing", "Sales"], 1000)
df["Gender"] = np.random.choice(["Male", "Female"], 1000)
df["Class"] = y   # Target variable

print("Dataset Shape:", df.shape)
print(df.head())

plt.figure(figsize=(14, 6))

# Numerical
plt.subplot(1, 2, 1)
sns.histplot(df["Age"], kde=True, bins=30, color="skyblue")
plt.title("Univariate: Age Distribution")

# Categorical
plt.subplot(1, 2, 2)
sns.countplot(x="Department", data=df, palette="Set2")
plt.title("Univariate: Department Count")

plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 6))

# Numerical vs Numerical
plt.subplot(1, 2, 1)
sns.scatterplot(x="Age", y="Salary", hue="Department", data=df, palette="Set1", alpha=0.7)
plt.title("Bivariate: Age vs Salary (by Department)")

# Numerical vs Categorical
plt.subplot(1, 2, 2)
sns.boxplot(x="Gender", y="Salary", data=df, palette="Set3")
plt.title("Bivariate: Salary by Gender")

plt.tight_layout()
plt.show()

# ---------------------------------
# Step 4: Multivariate Analysis
# ---------------------------------
# Pairplot (all numerical features with categorical hue)
sns.pairplot(df[["Age", "Salary", "Experience", "WorkingHours", "Projects", "PerformanceScore", "Department"]], 
             hue="Department", palette="Dark2")
plt.suptitle("Multivariate: Pair Plot", y=1.02)
plt.show()

# Correlation heatmap for numerical features
plt.figure(figsize=(10, 6))
sns.heatmap(df[["Age", "Salary", "Experience", "WorkingHours", "Projects", "PerformanceScore"]].corr(), 
            annot=True, cmap="coolwarm")
plt.title("Multivariate: Correlation Heatmap")
plt.show()

# Multivariate: Violin plot (salary distribution by department & gender)
plt.figure(figsize=(10, 6))
sns.violinplot(x="Department", y="Salary", hue="Gender", data=df, split=True, palette="muted")
plt.title("Multivariate: Salary Distribution by Department & Gender")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Number of samples
n = 1000

# Numerical features
age = np.random.normal(loc=35, scale=10, size=n).astype(int)
salary = np.random.normal(loc=70000, scale=15000, size=n).astype(int)
experience = np.clip(np.random.normal(loc=10, scale=5, size=n), 0, None).astype(int)
performance_score = np.random.uniform(1, 5, size=n)

# Categorical features
departments = ['Engineering', 'HR', 'Marketing', 'Sales', 'Finance', 'Product']
education_levels = ['High School', 'Bachelor', 'Master', 'PhD']
locations = ['New York', 'San Francisco', 'Austin', 'Seattle', 'Chicago']

department = np.random.choice(departments, size=n)
education = np.random.choice(education_levels, size=n, p=[0.1, 0.5, 0.3, 0.1])
location = np.random.choice(locations, size=n)

# Binary target variable
promotion = np.random.choice([0, 1], size=n, p=[0.8, 0.2])

# Create DataFrame
df = pd.DataFrame({
    'Age': age,
    'Salary': salary,
    'Experience': experience,
    'PerformanceScore': performance_score,
    'Department': department,
    'Education': education,
    'Location': location,
    'Promotion': promotion
})

print(df.head())


fig, axes = plt.subplots(2, 2, figsize=(12, 8))
sns.histplot(df['Age'], kde=True, ax=axes[0, 0])
sns.histplot(df['Salary'], kde=True, ax=axes[0, 1])
sns.histplot(df['Experience'], kde=True, ax=axes[1, 0])
sns.histplot(df['PerformanceScore'], kde=True, ax=axes[1, 1])
plt.tight_layout()
plt.show()

# Categorical
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Department')
plt.title('Department Distribution')
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Experience', y='Salary', hue='Promotion')
plt.title('Experience vs Salary')
plt.show()

# Categorical vs Numerical
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Education', y='Salary')
plt.title('Salary by Education Level')
plt.show()


# Pairplot with hue
sns.pairplot(df[['Age', 'Salary', 'Experience', 'PerformanceScore', 'Promotion']], hue='Promotion')
plt.show()

# Heatmap of correlations
plt.figure(figsize=(10, 6))
corr = df[['Age', 'Salary', 'Experience', 'PerformanceScore']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Categorical + Numerical + Target
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Department', y='PerformanceScore', hue='Promotion')
plt.title('Performance Score by Department and Promotion Status')
plt.xticks(rotation=45)
plt.show()

+
```
